```python
from pandas import read_csv
from matplotlib import pyplot

%matplotlib inline

filename = "data/pima-indians-diabetes.csv"
data = read_csv(filename)
```

* Age:                 Age (years)

* Mass BMI:            Body mass index (weight in kg/(height in m)^2)

* Pedi Diabetes:       Pedigree Function Diabetes pedigree function

* Plas Glucose:        Plasma glucose concentration a 2 hours in an oral glucose tolerance test

* Pregnancies:         Number of times pregnant

* Pres BloodPressure:  Diastolic blood pressure (mm Hg)

* Skin:                ThicknessTriceps skin fold thickness (mm)

* Test Insulin:       2-Hour serum insulin (mu U/ml)


* Class:               variable (0 or 1) 268 of 768 are 1, the others are 0

```python
#importante el figsize=(20,10) sino es minusculo
data.hist(figsize=(20,10))
pyplot.show()
```

```python
#importante el figsize=(20,10) sino es minusculo
data.plot(kind='density', subplots=True, layout=(3,3), figsize=(20,10), sharex=False)
pyplot.show()
```

```python
data.plot(kind='box', subplots=True, layout=(3,3), figsize=(20,10), sharex=False, sharey=False)
pyplot.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(data, figsize=(20,10))
pyplot.show()
```
