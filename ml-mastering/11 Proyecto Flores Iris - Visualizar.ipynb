```python
from pandas import read_csv
filename = 'data/iris.csv'
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = read_csv(filename, names=names)

dataset.head()
```

```python
dataset.describe()
```

```python
dataset.groupby('class').size()
```

```python
from matplotlib import pyplot

%matplotlib inline

#importante el figsize=(20,10) sino es minusculo
dataset.hist(figsize=(20,10))
pyplot.show()
```

```python
# box and whisker plots
dataset.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False, figsize=(20,10))
pyplot.show()
```

```python
import numpy
import pandas as pd

pd.set_option('display.width', 100)
pd.set_option('precision', 3)

correlations = dataset.corr(method='pearson')
correlations
```

```python
from matplotlib import pyplot

fig = pyplot.figure(figsize=(20,10))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)

ticks = numpy.arange(0,4,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
pyplot.show()
```

```python
from pandas.plotting import scatter_matrix
# scatter plot matrix
scatter_matrix(dataset, figsize=(20,10))
pyplot.show()
```
